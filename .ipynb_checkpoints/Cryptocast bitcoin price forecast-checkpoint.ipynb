```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
df.rename(columns={
    'Price': 'Close',
    'Vol.': 'Volume',
    'Change %': 'Change'
}, inplace=True)
```

```python
df
```

```python
for col in ['Close', 'Open', 'High', 'Low']:
    df[col] = df[col].str.replace(',', '').astype(float)
```

```python
df
```

```python
def convert_volume(val):
    if isinstance(val, str):
        val = val.replace(',', '')
        if 'K' in val:
            return float(val.replace('K','')) * 1e3
        elif 'M' in val:
            return float(val.replace('M','')) * 1e6
    return float(val)

df['Volume'] = df['Volume'].apply(convert_volume)
```

```python
df
```

```python
def convert_volume(val):
    if isinstance(val, str):
        val = val.replace(',', '').strip()
        
        if val == '-' or val == '':
            return None
        
        if 'K' in val:
            return float(val.replace('K','')) * 1e3
        elif 'M' in val:
            return float(val.replace('M','')) * 1e6
        elif 'B' in val:
            return float(val.replace('B','')) * 1e9
    
    return float(val)
```

```python
df['Volume'] = df['Volume'].apply(convert_volume)
```

```python
import numpy as np

def create_sequences(data, window=60):
    X, y1, y3, y7 = [], [], [], []

    for i in range(len(data) - window - 7):
        X.append(data[i:i+window])
        
        # Close price index = 3
        y1.append(data[i+window][3])      # 1-day
        y3.append(data[i+window+2][3])    # 3-day
        y7.append(data[i+window+6][3])    # 7-day

    return np.array(X), np.array(y1), np.array(y3), np.array(y7)

X, y1, y3, y7 = create_sequences(scaled_data)
```
